```python
import requests

# URL for updating user data (replace {id} with the user ID)
url = 'https://reqres.in/api/users/2'  # Example user ID to update

# Data to be sent in the PUT request
data = [
    {'name': 'john', 'job': 'Director'},
    {'name': 'Pater', 'job': 'CEO'},
    {'name': 'Lamin', 'job': 'Manager'}
]


# Step 1: Send a PUT request
response = requests.put(url, json=data)

# Step 2: Check if the request was successful
if response.status_code == 200:
    # Step 3: Access the response data
    print("Response from server:", response.json())
else:
    print(f"Failed to update data: {response.status_code}")
```

### PUT

```python
# URL for bulk updating users (replace with the actual endpoint if available)
r = 'https://reqres.in/api/users'  # Example endpoint

# Data to be sent in the bulk update request
data = [
    {'id': 298, 'name': 'john', 'job': 'Director'},
    {'id': 299, 'name': 'Pater', 'job': 'CEO'},
    {'id': 300, 'name': 'Lamin', 'job': 'Manager'}
]

# Sending the bulk update request
response = requests.put(url, json=data)  # Assuming the API accepts an array of users for bulk update

# Check if the request was successful
if response.status_code == 200:
    print("data updated successfully:", response.json())
else:
    print(f"Failed to update data: {response.status_code}, {response.text}")
```

### DELETE

```python
import requests

# URL for bulk deleting users
url = 'https://reqres.in/api/users'

# List of user IDs to be deleted
user_ids = [298, 299]

# Sending the bulk delete request
response = requests.delete(url, json={'ids': user_ids})

# Check if the request was successful
if response.status_code == 204:
    print("Users deleted successfully.")
else:
    print(f"Failed to delete users: {response.status_code}, {response.text}")
```

```python
tokens = [t for t in text.split()]
print(tokens)
```

```python
sr = stopwords.words('english')
clean_tokens = tokens[:]

for token in tokens:
  if token in stopwords.words('english'):
    clean_tokens.remove(token)
```

### Website to Scrape

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

# Initialize lists to store laptop data
laptop_name = []
laptop_rating = []
laptop_processor = []
laptop_memory = []
laptop_os = []
laptop_storage = []
laptop_display = []
laptop_inclusion = []
laptop_price = []

# Open the browser and navigate to the Flipkart page

driver.get('https://www.flipkart.com/search?q=gaming+laptop&sid=6bo%2Cb5g&sort=price_desc&page=1')

# Wait until the laptop elements are present
WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'CGtC98')))

laptops = driver.find_elements(By.CLASS_NAME, 'CGtC98')

for laptop in laptops:
    try:
        name = laptop.find_element(By.CLASS_NAME, 'KzDlHZ').text
        print(name)

        rating = laptop.find_element(By.CLASS_NAME, 'XQDdHH').text
        print(rating)

        price = laptop.find_element(By.CLASS_NAME, 'cN1yYO').text
        print(price)

        # Split the price lines if necessary
        lines_price = price.splitlines()
        if lines_price:
            discounted_price = lines_price[0]
            laptop_price.append(discounted_price)

        specs = laptop.find_element(By.CLASS_NAME, 'G4BRas').text
        print(specs)

        # Split specs into lines
        lines = specs.splitlines()
        if len(lines) >= 6:
            laptop_name.append(name)
            laptop_rating.append(rating)

            # Append specs
            laptop_processor.append(lines[0])
            laptop_memory.append(lines[1])
            laptop_os.append(lines[2])
            laptop_storage.append(lines[3])
            laptop_display.append(lines[4])
            laptop_inclusion.append(lines[5])

    except Exception as e:
        print(f"Error: {e}")  # Print the error for debugging

    sleep(2)  # Optional: Reduce the sleep time if necessary

# Optionally, close the driver after scraping
driver.quit()
```

### EDA

```python
# Check lengths of all lists
lengths = {
    'Name': len(laptop_name),
    'Price': len(laptop_price),
    'Rating': len(laptop_rating),
    'Processor': len(laptop_processor),
    'RAM': len(laptop_memory),
    'Operating System': len(laptop_os),
    'Storage': len(laptop_storage),
    'Display': len(laptop_display),
    'Inclusion': len(laptop_inclusion),
}

print(lengths)

# Ensure all lists have the same length before creating the DataFrame
if len(set(lengths.values())) == 1:  # Check if all lengths are the same
    data = {
        'Name': laptop_name,
        'Price': laptop_price,
        'Rating': laptop_rating,
        'Processor': laptop_processor,
        'RAM': laptop_memory,
        'Operating System': laptop_os,
        'Storage': laptop_storage,
        'Display': laptop_display,
        'Inclusion': laptop_inclusion
    }

    # Create the DataFrame
    df = pd.DataFrame(data)
    print(df.head())  # Display the first few rows of the DataFrame
else:
    print("Error: Lists are not of the same length. DataFrame cannot be created.")
```
